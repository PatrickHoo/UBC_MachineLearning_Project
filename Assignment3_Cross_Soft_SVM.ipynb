```python
# This is the cross-entropy function

from autograd import grad
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import datasets

import autograd.numpy as np

def model(x_p,w):
# compute linear comb and return
    a = w[0] + np.dot(x_p,w[1:])
 #   print("*******"+str(a))
    return a

def sigmoid(t):
    return 1/(1+np.exp(-t))

def cross_entropy(w,x,y):

    a = sigmoid(model(x,w ))
    # compute cost of label 0 points
    ind = np.argwhere(y==0)[:,0]
    
    cost = -np.sum(np.log(1-a[ind,:]))
    # add cost of label 0 points
    ind = np.argwhere(y==1)[:,0]
    #print(ind)
    cost -= np.sum(np.log(a[ind,:]))
    # compute cross entropy
    return cost/y.size

def gradient_descent(g, step, max_its ,w, p):
# compute gradient
    gradient = grad(g)
    
    weight_history = [w] # weight history
    cost_history = [g(w)] # cost history container
  #  print(max_its)

    for k in range(max_its):
        
      #  print("WW: "+str(w))
    #    print("hggg")
        grad_eval = gradient(w)
    #    print(grad_eval)
    #    print("***" + str(k))
     #   print(grad_eval_norm)
        grad_eval_norm = grad_eval/np.linalg.norm(grad_eval)
     #   print(grad_eval)
        
        if step == 'd': # diminishing step
            alpha = 1/(k+1)
        else: # constant step
            alpha = step

        w = w -alpha*grad_eval_norm


        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history , cost_history

iris = datasets.load_iris()
y = iris.target


x1 = iris.data[y<2,2]  # length?
x2 = iris.data[y<2,3]  # with?
y = y[y<2]
x=np.array([x1,x2])
x=x.T

#print(x.shape)
#print(x)
#print(y)

colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])
    
plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


#plt.plot(x1p,x2p)
#plt.xlim(1,5)
#plt.ylim(0,2)
def cross(t):
    cross =cross_entropy(t,x,y)
    return cross

iter =100
w = np.array ([[1.], [1.], [1.]])
#print(w)

a, b = gradient_descent(cross, 0.1, iter,w,0)
print(a[iter][1])



colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])

plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

x1p=np.linspace(1,5,20)
x2p=-(a[iter][0]+ a[iter][1]*x1p)/a[iter][2]

plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)


plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)
print(a[100])

#plt.plot(xp,sigmoid(model(xp,a[iter])))
```

```python
# This is the softmax function

from autograd import grad
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import datasets

import autograd.numpy as np

def model(x_p,w):
# compute linear comb and return
    a = w[0] + np.dot(x_p,w[1:])
 #   print("*******"+str(a))
    return a

def sigmoid(t):
    return 1/(1+np.exp(-t))

def softmax(w,x,y):
    
    
    y_temp = y*2-1

    a = -(y_temp.reshape(-1,1)*model(x,w )) 
    b = np.exp(a)+1
    

    
    cost = np.sum(np.log(b))

    
    return cost/y.size

def gradient_descent(g, step, max_its ,w, p):
# compute gradient
    gradient = grad(g)
    
    weight_history = [w] # weight history
    cost_history = [g(w)] # cost history container
  #  print(max_its)

    for k in range(max_its):
        
      #  print("WW: "+str(w))
    #    print("hggg")
        grad_eval = gradient(w)
    #    print(grad_eval)
    #    print("***" + str(k))
     #   print(grad_eval_norm)
        grad_eval_norm = grad_eval/np.linalg.norm(grad_eval)
     #   print(grad_eval)
        
        if step == 'd': # diminishing step
            alpha = 1/(k+1)
        else: # constant step
            alpha = step

        w = w -alpha*grad_eval_norm


        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history , cost_history

iris = datasets.load_iris()
y = iris.target


x1 = iris.data[y<2,2]  # length?
x2 = iris.data[y<2,3]  # with?
y = y[y<2]
x=np.array([x1,x2])
x=x.T

#print(x.shape)
#print(x)
#print(y)

colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])
    
plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


#plt.plot(x1p,x2p)
#plt.xlim(1,5)
#plt.ylim(0,2)
def soft(t):
    soft =softmax(t,x,y)
    return soft

iter =100
w = np.array ([[1.], [1.], [1.]])
#print(w)

a, b = gradient_descent(soft, 0.1, iter,w,0)
print(a[iter][1])



colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])

plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

x1p=np.linspace(1,5,20)
x2p=-(a[iter][0]+ a[iter][1]*x1p)/a[iter][2]

plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)


plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)
print(a[100])

#plt.plot(xp,sigmoid(model(xp,a[iter])))
```

```python
# This is the SVM function

from autograd import grad
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import datasets

import autograd.numpy as np

def model(x_p,w):
# compute linear comb and return
    a = w[0] + np.dot(x_p,w[1:])
 #   print("*******"+str(a))
    return a

def sigmoid(t):
    return 1/(1+np.exp(-t))

def Soft_margin_svm(w,x,y):
    
    
    y_temp = y*2-1
  #  print(softmax_y)
    #print(y.reshape(-1,1))
    a = -(y_temp.reshape(-1,1)*model(x,w )) +1
    b = np.exp(a)+1
    
    regular = 0.3*(np.linalg.norm(w[1:]))**2
    
    cost = np.sum(np.log(b)) + regular

    
    return cost/y.size

def gradient_descent(g, step, max_its ,w, p):
# compute gradient
    gradient = grad(g)
    
    weight_history = [w] # weight history
    cost_history = [g(w)] # cost history container
  #  print(max_its)

    for k in range(max_its):
        
      #  print("WW: "+str(w))
    #    print("hggg")
        grad_eval = gradient(w)
    #    print(grad_eval)
    #    print("***" + str(k))
     #   print(grad_eval_norm)
        grad_eval_norm = grad_eval/np.linalg.norm(grad_eval)
     #   print(grad_eval)
        
        if step == 'd': # diminishing step
            alpha = 1/(k+1)
        else: # constant step
            alpha = step

        w = w -alpha*grad_eval_norm


        weight_history.append(w)
        cost_history.append(g(w))
    return weight_history , cost_history

iris = datasets.load_iris()
y = iris.target


x1 = iris.data[y<2,2]  # length?
x2 = iris.data[y<2,3]  # with?
y = y[y<2]
x=np.array([x1,x2])
x=x.T

#print(x.shape)
#print(x)
#print(y)

colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])
    
plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


#plt.plot(x1p,x2p)
#plt.xlim(1,5)
#plt.ylim(0,2)
def svm_soft(t):
    svm_soft =Soft_margin_svm(t,x,y)
    return svm_soft

iter =100
w = np.array ([[1.], [1.], [1.]])
#print(w)

a, b = gradient_descent(svm_soft, 0.1, iter,w,0)
print(a[iter][1])



colors=('r','b')
for target in range(2):
    plt.scatter(x1[y ==target],x2[y==target],c=colors[target])

plt.grid(1)
plt.legend([iris.target_names[0], iris.target_names[1]])
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

x1p=np.linspace(1,5,20)
x2p=-(a[iter][0]+ a[iter][1]*x1p)/a[iter][2]

plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)


plt.plot(x1p,x2p)
plt.xlim(1,5)
plt.ylim(0,2)
print(a[100])

#plt.plot(xp,sigmoid(model(xp,a[iter])))
```
